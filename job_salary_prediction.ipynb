## 7. Outlier Detection Using IQR Method

```python
# Outlier Detection Using IQR Method

q1 = salary_df['salary'].quantile(0.25)
q3 = salary_df['salary'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

## 8. Salary Column Analysis Using NumPy Operations


```python
# Salary Column Operations

print("Salary Column Shape:", salary_df['salary'].shape)
print("Salary Data Type:", salary_df['salary'].dtype)
print("Total Salary Values:", salary_df['salary'].size)
print("Average Salary:", np.mean(salary_df['salary']))
print("Total Salary Sum:", np.sum(salary_df['salary']))
```

```python
# Salary Column NumPy Operations

salary_array = np.array(salary_df['salary'])

print("Salary NumPy Array:", salary_array)
print("Array Shape:", salary_array.shape)
print("Array Data Type:", salary_array.dtype)
print("First 5 Salary Values:", salary_array[:5])
print("Maximum Salary:", np.max(salary_array))
print("Minimum Salary:", np.min(salary_array))
print("Average Salary:", np.mean(salary_array))
print("Total Salary:", np.sum(salary_array))
```

## 9. Comprehensive Salary Analysis Using Multiple Visualizations

```python
# Comprehensive Salary Analysis Using Multiple Visualizations

fig, axes = plt.subplots(2, 3, figsize=(20, 15))

fig.suptitle('Job Salary Analysis', fontsize=16, fontweight='bold')


# 1. Top 20 Highest Salaries
top_salary = salary_df.nlargest(20, 'salary')[['job_title', 'salary']]

axes[0, 0].barh(top_salary['job_title'], top_salary['salary'],
                color='skyblue', alpha=0.7)

axes[0, 0].set_title('Top 20 Highest Salaries', fontweight='bold')
axes[0, 0].set_xlabel('Salary')
axes[0, 0].grid(axis='x', alpha=0.3)


# 2. Salary Distribution
axes[0, 1].hist(salary_df['salary'], bins=20)

axes[0, 1].set_title('Salary Distribution', fontweight='bold')
axes[0, 1].set_xlabel('Salary')
axes[0, 1].set_ylabel('Frequency')


# 3. Experience vs Salary
axes[0, 2].scatter(salary_df['experience_years'], salary_df['salary'], alpha=0.6)

axes[0, 2].set_title('Experience vs Salary', fontweight='bold')
axes[0, 2].set_xlabel('Experience Years')
axes[0, 2].set_ylabel('Salary')
axes[0, 2].grid(True, alpha=0.3)


# 4. Education Level Analysis
education_salary = salary_df.groupby('education_level')['salary'].mean()

axes[1, 0].bar(education_salary.index, education_salary.values,
               color='lightcoral', alpha=0.7)

axes[1, 0].set_title('Average Salary by Education Level', fontweight='bold')
axes[1, 0].set_xlabel('Education Level')
axes[1, 0].set_ylabel('Average Salary')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(axis='y', alpha=0.3)


# 5. Industry Wise Salary
industry_salary = salary_df.groupby('industry')['salary'].mean().sort_values(ascending=False)

axes[1, 1].barh(industry_salary.index, industry_salary.values,
                color='lightgreen', alpha=0.7)

axes[1, 1].set_title('Average Salary by Industry', fontweight='bold')
axes[1, 1].set_xlabel('Average Salary')
axes[1, 1].grid(axis='x', alpha=0.3)


# 6. Remote Work Distribution
remote_count = salary_df['remote_work'].value_counts()

axes[1, 2].pie(remote_count,
               labels=remote_count.index,
               autopct='%1.1f%%',
               startangle=90)

axes[1, 2].set_title('Remote Work Distribution', fontweight='bold')


plt.tight_layout()

plt.show()
```

## 10. Salary Analysis Using GroupBy and Line Plot

```python
# Salary Analysis Using GroupBy and Line Plot

grouped_data = salary_df.groupby("education_level")["salary"].mean()

result = pd.DataFrame({"Average Salary": grouped_data})

plt.plot(result, c="green", marker="o")

plt.xlabel("Education Level", c="red", size=20)

plt.ylabel("Average Salary", c="red", size=20)

plt.title("Average Salary by Education Level", size=20)

plt.grid(True)

plt.show()
```
